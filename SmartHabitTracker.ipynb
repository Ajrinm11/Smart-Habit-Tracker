```python
class Habit:
    def __init__(self, habitName, streak=0, donetoday=False):
        self.name = habitName
        self.streak = streak
        self.completed = donetoday

    def markasdone(self):
        if not self.completed:
            self.completed = True
            self.streak += 1
            print(f"âœ… Great job! You completed {self.name}! ğŸ”¥ Your streak is now {self.streak}.")

    def notdonetoday(self):
        if not self.completed:
            print(f"âŒ You missed {self.name} today! Your streak has reset back to 0.")
            self.streak = 0
        self.completed = False

    def status(self):

        print(f"\nğŸ“‹ Habit: {self.name}")
        print(f"âœ… Done today: {'Yes' if self.completed else 'No'}")
        print(f"ğŸ”¥ Current streak: {self.streak}")

        if self.streak == 0:
            print("â˜€ï¸ Let's get started! Every journey begins with a single step.")
        elif self.streak < 5:
            print("ğŸ’ª Keep it up! You're building up pace!")
        elif self.streak < 10:
            print("ğŸš€ You're on fire! Stay consistent!")
        else:
            print("ğŸ† You're a pro! Amazing dedication!")


habit_list = []

def addhabit():
  name = input("Enter new habit: ")
  habitobj = Habit(name)
  habit_list.append(habitobj)
  print(f"â˜‘ï¸ Habit {name} added!\n")


def showAllHabits():
  if len(habit_list)==0:
    print("No habits added yet!\n")
    return
  for index, item in enumerate(habit_list):
    print(f"{index+1}. {item.name}\n")
  print()

def selecthabit():
  showAllHabits()
  if len(habit_list)==0:
    return None
  try:
    habitselected = int(input("Select a habit by number: "))-1
    if 0 <= habitselected < len(habit_list):
      return habit_list[habitselected]
    else:
      print("Invalid choice!\n")
      return None

  except ValueError:
    print("Enter a valid number\n")
    return None

def nextdayAll():
  for h in habit_list:
    h.notdonetoday()
  print("â˜€ï¸ Welcome to a new day! All statuses are set to 0.\n")


def mainmenu():
  while True:
    print()
    print("ğŸƒğŸª´ğŸƒğŸª´ğŸƒğŸª´ğŸƒğŸª´ğŸƒğŸª´ğŸƒ")
    print("â­ --- Habit Hero Menu --- â­")
    print("1. Add new habit")
    print("2. Mark habit as complete")
    print("3. Show habit status")
    print("4. Move to next day")
    print("5. Quit.")
    print("ğŸƒğŸª´ğŸƒğŸª´ğŸƒğŸª´ğŸƒğŸª´ğŸƒğŸª´ğŸƒ")
    print()

    choice = input("Choose a valid option (1-5): ")
    print()
    if choice == '1':
      addhabit()

    elif choice == '2':
      select = selecthabit()
      if select:
        select.markasdone()

    elif choice == '3':
      print("â˜ï¸â„ï¸â˜ï¸â„ï¸â˜ï¸â„ï¸â˜ï¸â„ï¸â˜ï¸â„ï¸â˜ï¸")
      print("â­ --- Current Habits' Status --- â­")
      for h in habit_list:
        h.status()
      print("â˜ï¸â„ï¸â˜ï¸â„ï¸â˜ï¸â„ï¸â˜ï¸â„ï¸â˜ï¸â„ï¸â˜ï¸")

    elif choice == '4':
      nextdayAll()

    elif choice == '5':
      print("ğŸ‘‹ Byeeee! Keep striving with your habitssss!")
      break

    else:
      print("Invalid option! Please choose a valid option from 1-5.")


```
